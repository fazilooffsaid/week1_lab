```python
text = "GenAI is amazing! GenAI helps students, and GenAI can write code."
text = text.lower()
punctuation = ".,!?"
for p in punctuation:
    text = text.replace(p, "")
words = text.split()
word_freq = {}

for word in words:
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1
print(word_freq)
```

Problem 6



```python
import math
a = [1, 2, 3]
b = [2, 4, 1]
def dot_product(v1, v2):
    result = 0
    for x, y in zip(v1, v2):
        result += x * y
    return result
def cosine_similarity(v1, v2):
    dot = dot_product(v1, v2)
    norm_v1 = math.sqrt(dot_product(v1, v1))
    norm_v2 = math.sqrt(dot_product(v2, v2))
    return dot / (norm_v1 * norm_v2)
print("Dot product:", dot_product(a, b))
print("Cosine similarity:", cosine_similarity(a, b))
```

Problem 7


```python
import math
scores = [2.0, 1.0, 0.1]
def softmax(scores):
    exp_scores = []
    for s in scores:
        exp_scores.append(math.exp(s))
    total = sum(exp_scores)
    probs = []
    for e in exp_scores:
        probs.append(e / total)
    return probs
probabilities = softmax(scores)
print(probabilities)
print("Sum:", sum(probabilities))
```

Problem 8


```python
w = [0.2, -0.1]
b = 0.0
x = [1.0, 2.0]
y_true = 1
lr = 0.5
z = w[0]*x[0] + w[1]*x[1] + b
y_hat = 1 if z >= 0 else 0
print("Prediction y_hat:", y_hat)
w[0] = w[0] + lr * (y_true - y_hat) * x[0]
w[1] = w[1] + lr * (y_true - y_hat) * x[1]
b = b + lr * (y_true - y_hat)
print("Updated w:", w)
print("Updated b:", b)
```

Problem 9


```python
X = [[1,1],[2,1],[2,3],[3,3]]
y = [0,0,1,1]

w = [0,0]
b = 0
lr = 0.1

for _ in range(10):
    for x, yt in zip(X, y):
        yh = 1 if (w[0]*x[0] + w[1]*x[1] + b) >= 0 else 0
        w[0] += lr*(yt-yh)*x[0]
        w[1] += lr*(yt-yh)*x[1]
        b += lr*(yt-yh)

correct = sum(
    1 for x, yt in zip(X, y)
    if (1 if (w[0]*x[0] + w[1]*x[1] + b) >= 0 else 0) == yt
)

print("w:", w)
print("b:", b)
print("accuracy:", correct/len(y))
```

Problem 10


```python
sentences = [
    "genai is fun",
    "genai is useful",
    "ai is useful",
    "genai can write code",
    "ai can write code"
]
bigram_counts = {}
for sentence in sentences:
    words = sentence.split()
    for i in range(len(words) - 1):
        w1, w2 = words[i], words[i+1]

        if w1 not in bigram_counts:
            bigram_counts[w1] = {}

        bigram_counts[w1][w2] = bigram_counts[w1].get(w2, 0) + 1
def predict_next(word):
    if word not in bigram_counts:
        return None
    return max(bigram_counts[word], key=bigram_counts[word].get)
print("Bigram counts:")
for k, v in bigram_counts.items():
    print(k, "->", v)
print("\nPredictions:")
print("genai →", predict_next("genai"))
print("is →", predict_next("is"))
print("can →", predict_next("can"))
```
